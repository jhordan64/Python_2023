   - Muestre estas medidas de tendencia central en un BoxPlot para cada variable mencionada.

```python
def boxplot_crear(data, column):
    sns.boxplot(
        data=data,
        x=column,
        color='yellow'  # Cambiar el color a amarillo
    )

    plt.axvline(
        x=data[column].mean(),
        color='red',
        linestyle='dashed',
        linewidth=2
    )
    plt.show()
```

```python
# Crear boxplots para cada variable con la función
boxplot_crear(bbdd_df, 'MORA60')
boxplot_crear(bbdd_df, 'ATRASO')
boxplot_crear(bbdd_df, 'EDAD')
boxplot_crear(bbdd_df, 'TC')
boxplot_crear(bbdd_df, 'AHORRO')
boxplot_crear(bbdd_df, 'INGRESO')
boxplot_crear(bbdd_df, 'LINEA_TC')
boxplot_crear(bbdd_df, 'SALDO_DEUDOR')
boxplot_crear(bbdd_df, 'SCORE')
boxplot_crear(bbdd_df, 'ABONO_PAGO')
boxplot_crear(bbdd_df, 'RIESGO_SBS')
```

## 4. Visualización de Datos

  - Genere un gráfico de barras para todas las variables numéricas.

```python
def barras_crear(data, column):
    sns.catplot(
       data=data,
      x=column,
       kind='count'
    )
    plt.xticks(rotation=90)
    plt.show()
```

```python
# Crear barras para cada variable con la función
barras_crear(bbdd_df, 'MORA60')
barras_crear(bbdd_df, 'ATRASO')
barras_crear(bbdd_df, 'EDAD')
barras_crear(bbdd_df, 'TC')
barras_crear(bbdd_df, 'AHORRO')
barras_crear(bbdd_df, 'INGRESO')
barras_crear(bbdd_df, 'LINEA_TC')
barras_crear(bbdd_df, 'SALDO_DEUDOR')
barras_crear(bbdd_df, 'SCORE')
barras_crear(bbdd_df, 'ABONO_PAGO')
barras_crear(bbdd_df, 'RIESGO_SBS')
```

## 6. Análisis de Variable Objetivo

   - ¿Qué porcentaje de clientes ha tenido una morosidad mayor a 60 días (`MORA60`) después de la aprobación del crédito?

```python
#Porcentaje
morosidad = bbdd[bbdd['ATRASO'] > 60]

porcentaje = (len(morosidad) / len(bbdd)) * 100

print(f"Porcentaje de clientes con morosidad mayor a 60 días: {porcentaje:.2f}%")


#Conteo
cont = bbdd[bbdd['ATRASO'] > 60]

cantidad_morosos = len(cont)
cantidad_morosos
```

   - Investigue si existe alguna relación entre el nivel educativo y la incidencia de morosidad mayor a 60 días. ¿Cómo podría visualizar esta relación?

```python
tab = pd.crosstab(bbdd['MORA60'],bbdd['EDUCACION'])

sns.heatmap(tab)
plt.show()
```
